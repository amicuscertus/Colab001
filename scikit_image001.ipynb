```
from skimage import data, io, filters, color
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```
url = 'https://github.com/amicuscertus/Test001/raw/master/sweets.jpg'
image = io.imread(url)
io.imshow(image)
```

```
url = 'https://github.com/amicuscertus/Test001/raw/master/celine-dion.jpg'
celine = io.imread(url)
io.imshow(celine)
```

```
celine_gray = rgb2gray(celine)
io.imshow(celine_gray)
```

```
red_multiplier = [1, 0, 0]
yellow_multiplier = [0, 1, 1]

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax1.imshow(red_multiplier * celine)
ax2.imshow(yellow_multiplier * celine)
```

```
hue_gradient = np.linspace(0, 1, 100)
hsv = np.ones(shape=(1, len(hue_gradient), 3), dtype=float)
hsv[:, :, 0] = hue_gradient
all_hues = color.hsv2rgb(hsv)
fig, ax = plt.subplots(figsize=(10, 2))
ax.imshow(all_hues, extent=(0, 1, 0, 0.2))
```

```
celine_hsv = color.rgb2hsv(celine)
value = 0.0
saturation = 0.0
hue = 0.09
celine_hsv[:, :, 2] += value
celine_hsv[:, :, 1] += saturation
celine_hsv[:, :, 0] += hue
celine_new = color.hsv2rgb(celine_hsv)
plt.imshow(celine_new)
```

```
r = celine[:, :, 0].reshape(celine.size//3)
g = celine[:, :, 1].reshape(celine.size//3)
b = celine[:, :, 2].reshape(celine.size//3)
plt.hist([r, g, b], bins=16, range=(0, 255), color=['r', 'g', 'b'], stacked=False)
```

```
plt.hist(r, bins=32, color='r', alpha=0.5)
plt.hist(g, bins=32, color='g', alpha=0.5)
plt.hist(b, bins=32, color='b', alpha=0.5)
```

```
df = pd.DataFrame()
df['R'] = r
df['G'] = g
df['B'] = b
df.head(3)
```

```
sns.pairplot(df)
```

```
sns.distplot(r, color='r')
sns.distplot(g, color='g')
sns.distplot(b, color='b')
```
